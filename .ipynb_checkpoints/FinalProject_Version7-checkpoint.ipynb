## 1.2 Extracting the Data

There are 3 different sets of data that we will be extracting for this project:
 - Data Set 1: Player's and their Statistics (main Data Set)
 - Data Set 2: NBA 2K Player Ratings
 - Data Set 3: Player All Star Votes

#### Data Set 1: Player's and their Statistics:
 - This Data Set contains all the NBA players from the 2018-2019 NBA Season
 - For each player, it contains basic attributes for the player such as Team, Height, Weight, Age, etc.
 - Along with the attrivutes, it contains the players averages for each game of the season, such as Point, Rebounds, Assists, Steals, Field Goal Percentage, etc.
 - Link to the Kaggle Data Set is here: https://www.kaggle.com/mcamli/nba17-18#nba_extra.csv. Important thing to note, the name of the file says "2017-2018" but it is really the 2018-2019 Season. The file is nba_extra.csv

```python
NBA_Players = pd.read_csv("nbastats2018-2019.csv", sep=',')
NBA_Players = NBA_Players.dropna()

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
pd.set_option('display.max_colwidth', -1) 
NBA_Players.head(10)
```

#### Data Set 2: NBA 2K Player Ratings:
 - These are Player ratings taken from the NBA based video game NBA 2K19
 - These ratings are from 0 (being the worst) to 99 (being the best)
 - The ratings are based by predetermined attributes that are not depicted in the data set
 - The purpose of these ratings is to give a quantitive value to how good a player is, so we can order the players from best to worst if need be.
 - Link to the Kaggle Data Set is here: https://www.kaggle.com/stormex/nba-2019-ratings

```python
Ratings = pd.read_csv("Nba 19 ratings.csv", sep=',')
Ratings = Ratings.loc[:, ~Ratings.columns.str.contains('^Unnamed')]
Ratings.columns = ['Name', 'Rating']
Ratings.head(10)
```

#### Data Set 3: Player All Star Votes
 - This data is scraped off of the website https://www.basketball-reference.com/allstar/NBA_2019_voting-backcourt-eastern-conference.html. 
 - On the website there are four tabs. Two for each conference and in each conference it is split between forwards and guards
 - Each data set contains votes from fans, players, and media. We will be using the player votes.
 - In this data set, we will also get which conference (East or West) and which position each player is.

```python
urls = [["https://www.basketball-reference.com/allstar/NBA_2019_voting-frontcourt-eastern-conference.html","East","Forward"],
       ["https://www.basketball-reference.com/allstar/NBA_2019_voting-frontcourt-western-conference.html","West","Forward"],
       ["https://www.basketball-reference.com/allstar/NBA_2019_voting-backcourt-eastern-conference.html","East","Guard"],
       ["https://www.basketball-reference.com/allstar/NBA_2019_voting-backcourt-western-conference.html","West","Guard"]]

players = []
votes = []
position = []
conference = []
for val in urls:
    url = val[0]
    conf = val[1]
    pos = val[2]
    r = requests.get(url).text
    soup = BeautifulSoup(r ,'lxml')
    table = soup.find('table')
    list_of_rows = []
    for row in table.findAll('tr'):
        list_of_cells = []
        for cell in row.findAll(["th","td"]):
            text = cell.text
            list_of_cells.append(text)
        list_of_rows.append(list_of_cells)
    list_of_rows.pop(0)
    list_of_rows.pop(0)
    
    for player in list_of_rows:
        players.append(player[1])
        votes.append(player[5])
        conference.append(conf)
        position.append(pos)

Player_Votes = pd.DataFrame()
Player_Votes['Name'] = players
Player_Votes['Votes'] = votes
Player_Votes['Position'] = position
Player_Votes['Conference'] = conference
Player_Votes.head(10)
```

## 1.3 Merging the Data

Now that we have the three data sets, we can merge them all together on the column 'Name'.
We will also sort the values by votes.

```python
Complete_Data = pd.merge(NBA_Players, pd.merge(Ratings, Player_Votes, on='Name'), on='Name')
Complete_Data.Votes = Complete_Data.Votes.astype(float)
Complete_Data = Complete_Data.sort_values(by = ['Votes'], ascending = False)
Complete_Data.head(10)
```

# 4. Linear Regression for Attributes on Rating and Votes

Here, there is a comparison between the Ratings of the top 200 Players to the Votes of the top 200 Players. Both utilize attributes regarding player performances. The ratings from NBA 2k19 game are based on player skills. Popular beliefs would point towards the most skilled players being the most popular players. The Linear Regression Analysis below was done to determine whether or not there is a dichotomy between popularity (vote results) and skill rating.

## 4.1 Regression 

### Linear Regression Model for Rating:
The five Skill Sets: Points, Rebounds, Assists, Blocks, Steals per game do not affect the Rating.
Independent Variables: Points, Rebounds, Assists, Blocks, Steals
Dependent Variable: Rating
 

```python
skills = ['Points','Rebounds','Assists','Blocks','Steals']
player_stats = Top_200_Players[skills]
rating = Top_200_Players[['Rating']]

X = player_stats
y = rating['Rating']
lm = LinearRegression()
model = lm.fit(X, y)
lm.score(X, y)
```

The R-Square for this 0.91. This means that most of the variance for Rating values is explained by this model.

```python
rating_coeffs = lm.coef_.tolist()
for skill, coef in zip(skills, rating_coeffs):
    print("Skill: {}, Coefficient: {}".format(skill,coef))
```

From the above, it seems that the most important attributes are points per game and blocks per game. 

### Linear Regression Model for Votes:
The five skill sets: Points, Rebounds, Assists, Blocks, Steals per game do not affect the Rating.
Independent Variables: Points, Rebounds, Assists, Blocks, Steals
Dependent Variable: Votes

```python
skills2 = ['Points','Rebounds','Assists','Blocks','Steals']
player_stats2 = Top_200_Players[skills2]
X2 = player_stats2
votes = Top_200_Players[['Votes']]
y2 = votes['Votes']
lf = LinearRegression()
model = lf.fit(X2, y2)
lf.score(X2, y2)
```

The R-Square for this is 0.42. This means that not a whole lot of Voting values is explained by this model. This indicates there may perhaps be other factors besides players statistics regarding the number of votes.

```python
voting_coeffs = lf.coef_.tolist()
for skill, coef in zip(skills2, voting_coeffs):
    print("Attribute: {}, Coefficient: {}".format(skill,coef))
```

From the above, it seems that the most important attributes are also points per game and blocks per game. 

## 4.2 Prediction of Ratings vs Prediction of Votes (Machine Learning)

Here, we predict Ratings and Votes by splitting and training our data to create predictions. We use a test_size of 0.2 for both Ratings and Votes.

## Creating Predictions for Ratings

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
rating_regressor = LinearRegression()
prediction_model = rating_regressor.fit(X_train, y_train)
predictions = rating_regressor.predict(X_test)
plt.title("Predicted vs Actual Ratings for Top 200 Players")
plt.scatter(y_test, predictions)
plt.plot(y_test, y_test, color="green")
plt.xlabel("Actual Ratings")
plt.ylabel("Predicted Ratings")
```

```python
print("On scale from 0 to 1, the number to the right is how accurate the model predicts Ratings: ", prediction_model.score(X_test, y_test))
```

As you can observe, the model predicted the Top Player ratings rather well.

## Creating Predictions for Votes

```python
X_train, X_test, y_train, y_test = train_test_split(X2, y2, test_size=0.2, random_state=0)
rating_regressor = LinearRegression()
prediction_model = rating_regressor.fit(X_train, y_train)
predictions = rating_regressor.predict(X_test)
plt.title("Predicted vs Actual Votes for Highest Top 200 Players")
plt.scatter(y_test, predictions)
plt.plot(y_test, y_test, color="green")
plt.xlabel("Actual Votes")
plt.ylabel("Predicted Votes")
```

Here you can observe that the model does not seem to accurately predict number of Votes for each player.

```python
print("On scale from 0 to 1, the number to the right is how accurate the model predicts Votes: ", prediction_model.score(X_test, y_test))
```
